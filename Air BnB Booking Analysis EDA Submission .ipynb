#### Chart - 3

````python
# To remove outliers from the boxplot, you can use statistical methods such as the interquartile range (IQR)
# Calculate the IQR
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to exclude outliers
filtered_df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

# Create the boxplot without outliers
sns.boxplot(x='room_type', y='price', data=filtered_df)
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.title('Boxplot of Airbnb Prices by Room Type (Outliers Removed)')
plt.show()
````

##### 1. Why did you pick the specific chart?

1-Boxplots are effective at identifying outliers in a dataset. By plotting the data in a box-and-whisker format, outliers are shown as individual points outside the whiskers, making them visually distinct.
2- Boxplots provide summary statistics such as the median (the line inside the box), quartiles (the edges of the box), and potential outliers, giving a quick overview of the data's central tendency and spread.
3- Boxplots can reveal information about the skewness and symmetry of the data. For example, a skewed distribution may have a longer whisker on one side.

##### 2. What is/are the insight(s) found from the chart?

1- The boxplot allows you to visually compare the distribution of Airbnb prices for different room types. You can see the range of prices for each room type, as indicated by the whiskers, and the spread of prices within the boxes.
2-  The boxplot shows the median of the prices for each room type, represented by the line inside the box. The median gives an idea of the central tendency of the prices in each category.
3- By comparing the boxplots side by side, you can identify differences in the price distributions between different room types. For example, you can observe if certain room types tend to have higher or lower prices compared to others.

The specific chart chosen is the "Heatmap of Correlation Matrix" because it is an effective way to visualize the correlation between different numerical variables in the Airbnb dataset

##### 2. What is/are the insight(s) found from the chart?

Correlation Strength: The intensity of the color in the heatmap represents the strength of correlation between variables. Darker colors (either dark blue or dark red) indicate stronger positive or negative correlations, respectively.

Positive Correlations: Positive correlations are shown by dark blue squares. Variables with positive correlations tend to increase or decrease together. For example, there may be a positive correlation between the number of minimum_nights and the price of listings.

Negative Correlations: Negative correlations are shown by dark red squares. Variables with negative correlations move in opposite directions. For example, there might be a negative correlation between the longitude and the price of listings.

Strong Correlations: The heatmap highlights variables that have strong correlations with each other, helping to identify important relationships in the dataset.

No Correlation: Light-colored squares indicate little or no correlation between variables. These variables are likely to be independent of each other.
